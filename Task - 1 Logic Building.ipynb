# 1. USER LOGIN CHECK

```python
username = "admin"
password = "1234"

input_username = input("Enter username:") # taking inputs
input_password = input("Enter password:")

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer

```python
#Given marks
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0
#checking student marks
for m in marks:
    if m>=50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1

# printing final result
print("Total Students:", len(marks))
print("Number of Students Passed:", pass_count)
print("Number of Students Failed:", fail_count)
```

# 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]

Cleaned_names = []

#cleaning names
for n in names:
    name = n.strip() #remove extra space
    name = name.lower() # convert to lowercase
    Cleaned_names.append(name)

#printing the final list
print("Original List:", names)
print("Cleaned List:", Cleaned_names)
```

# 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

#checking each message
for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)
    if length > 10:
        print("This message is longer than 10 characters")
    print()
        
```

# 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count = error_count +1

print("Total ERROR entries:", error_count) # print final result
```

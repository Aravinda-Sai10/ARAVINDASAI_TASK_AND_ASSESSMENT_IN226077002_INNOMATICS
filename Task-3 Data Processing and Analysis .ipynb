# 1. Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# find highest score
max_score = 0

for name in employees:
    if employees[name] > max_score:
        max_score = employees[name]

# finding all employees with highest score
top_performers = []

for name in employees:
    if employees[name] == max_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", max_score, ")")
```

# 2. Search Query Keyword Analysis

```python
text = "Buy mobile phone buy phone online"

# converting to lowercase
text = text.lower()

# remove punctuation
text = text.replace(",", "").replace(".", "").replace("!", "")

# split into words
words = text.split()

# count frequency
freq = {}

for w in words:
    if w in freq:
        freq[w] = freq[w] + 1
    else:
        freq[w] = 1

# print only words with frequency > 1
result = {}

for key in freq:
    if freq[key] > 1:
        result[key] = freq[key]

print(result)
```

# 3. Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# checking each reading with index
for i in range(len(sensor_readings)):
    value = sensor_readings[i]
    
    # checking if even
    if value % 2 == 0:
        valid_readings.append((i, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# 4. Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# count domains
for mail in emails:
    parts = mail.split("@")
    domain = parts[1]
    
    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1

total = len(emails)

# calculate percentage
for domain in domain_count:
    percent = (domain_count[domain] / total) * 100
    print(domain + ":", int(percent), "%")
```

# 5. Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# finding average sales
total = 0
for s in sales:
    total = total + s

avg = total / len(sales)

# check for spikes 
for i in range(len(sales)):
    if sales[i] > avg * 1.3:
        print("Day", i + 1, ":", sales[i])
```

# 6. Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

# count frequency of each user id
for uid in user_ids:
    if uid in count:
        count[uid] = count[uid] + 1
    else:
        count[uid] = 1

# print only duplicates
for key in count:
    if count[key] > 1:
        print(key, "→", count[key], "times")
```

# 1. Smart Light Controller

```python
class SmartLight:
    
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    # Method to turn ON light
    def turn_on(self):
        self.status = "ON"
    
    # Method to turn OFF light
    def turn_off(self):
        self.status = "OFF"
    
    # Method to display status
    def show_status(self):
        print(self.name, "is", self.status)

light_name = "Bedroom Light"
action = "ON"

# Create object
light = SmartLight(light_name)

# Perform action
if action == "ON":
    light.turn_on()
else:
    light.turn_off()

# Display result
light.show_status()
```

# 2. Employee ID Card System


```python
class Employee:
    
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    # Method to display ID card
    def show_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample input
emp_name = "Rahul"
emp_id = "EMP102"
dept = "IT"

# Create object
emp = Employee(emp_name, emp_id, dept)

emp.show_id_card()
```

# 3. Mobile Contact Record

```python
class Contact:
    
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # Method to display contact info
    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample input
contact_name = "Anita"
phone_number = "9876543210"

# Create object
c1 = Contact(contact_name, phone_number)

# Display contact
c1.show_contact()
```

# 4. Product Price Tag Generator


```python
class Product:
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    # Method to print price tag
    def show_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


product_name = "Headphones"
product_price = 2499

# Create object
p1 = Product(product_name, product_price)

# Display price tag
p1.show_price_tag()
```



# 5. Movie Rating Display System

```python
class Movie:
    
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    # Method to display movie details
    def show_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


movie_name = "Inception"
movie_rating = 4.8

# Create object
m1 = Movie(movie_name, movie_rating)

# Display details
m1.show_details()
```

# 6. Delivery Address Manager


```python
class Delivery:
    
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    # Method to display delivery details
    def show_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


name = "Suman"
addr = "Hyderabad"

# Create object
d1 = Delivery(name, addr)

# Display delivery info
d1.show_delivery()
```
